```python
import os
import sys
import re
import gzip

filename_tbl = 'Suweon_homo.txt_fission' #sys.argv[1]
filename_tbl2 = 'mart_export(1).txt' #sys.argv[2]
filename_tbl3 = 'NIBR201605_SuweonTreeFrog109176_tx.combined_nTx_NR_prot6withposition.fa' #sys.argv[3]

f_dict1 = {}
with open(filename_tbl,"r") as f:
    for line in f:
        list1 = line.split()
        pid = list1[0]
        tid = list1[3]
        plen = list1[1]
        alignlen = list1[5]
        miss = list1[6]
        evalue = list1[-1]
        f_dict1[tid+" "+pid] = [plen,alignlen,miss,evalue]

'''
#when we have to use fasta file
#(choose only odd_number line)

f_dict2 = {}
with open(filename_tbl2,"r") as f:
    for i, line in enumerate(f):
        if i % 2 == 0:
            list1 = line.split()
            tid = list1[0]
            pid = list1[1]
            if len(list1) > 2:
                gid = list1[2]
            else : gid = None
        else :
            seq = line
            f_dict2[tid] = [pid, gid, seq]
'''
f_dict2 = {}
with open(filename_tbl2,"r") as f:
    for line in f:
        list1 = line.split()
        if len(list1) > 2:
            pid = list1[1]
            gid = list1[2]
        else :
            pid = list1[1]
            gid = ""
        f_dict2[pid] = [gid]

f_dict3 = {}
with open(filename_tbl3,"r") as f:
    for line in f:
        list1 = line.split()
        tid = list1[0][1:]
        position = list1[1]
        seq = list1[2]
        f_dict3[tid] = [position,seq]

with open('tested.fa','w') as f_p:
    for tid_pid in f_dict1.keys():
        tid, pid = tid_pid.split()
        if pid in f_dict2 and tid in f_dict3:
            
            plen =      f_dict1[tid_pid][0]
            alignlen =  f_dict1[tid_pid][1]
            miss =      f_dict1[tid_pid][2]
            align_per = (float(alignlen)-float(miss))/float(plen)
            evalue =    f_dict1[tid_pid][3]
            
            gid = f_dict2[pid][0]
            
            seq =      f_dict3[tid][1]
            position = f_dict3[tid][0]
            
            if float(alignlen) > 0.5*float(plen) and float(alignlen) > 0.5*len(seq):
                realign = float(alignlen)-float(miss)
                if float(realign) > 0.25*float(plen) and float(realign) > 0.25*len(seq):
                    f_p.write(tid+'\t'+gid+'\t'+str(round(align_per*100,1))+'%\t'+position+'\t'+evalue+'\t'+seq+'\n')
```
