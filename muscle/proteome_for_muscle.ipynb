```python
import os
import sys

f_selected = open('rm_noname.txt_rm_duplication_1to1','r')
f_tro = open('trop.fasta','r')
f_chi = open('chicken.fasta','r')
f_hum = open('human.fasta','r')
f_mus = open('mus.fasta','r')
f_my = open('processed.fa','r')

try:
    os.makedirs("my")
    os.makedirs("chicken")
    os.makedirs("mus")
    os.makedirs("trop")
    os.makedirs("human")
except: pass

gname = []
for line in f_selected:
    gname.append(line.split()[1])
f_selected.close()
a = list(set(gname))
a.sort()

my = {}
for line in f_my:
    if line.startswith('>'):
        splt = line.split('|')
        gname = splt[2].strip()
        my[gname] = ''
    else:
        my[gname] += line.strip()
        
def make_dict(species_name,filename):
    dict1 = {}
    for line in filename:
        if line.startswith('>'):
            splt = line.split()
            num1 = 0
            for i in splt:
                if i.find('GN') != -1:
                    g = i[3:]
                    gname = g.upper()
                    dict1[gname] = ''
                    num1 = 1
        else:
            if num1 == 1:
                dict1[gname] += line.strip()
    return dict1

tro = make_dict('tro',f_tro)
chi = make_dict('chi',f_chi)
mus = make_dict('mus',f_mus)
hum = make_dict('hum',f_hum)

def make_files(directory,dictname,a):
    b = []
    output_path = directory + '/out'
    for key in a:
        if key in dictname:
            b.append(key)
            f_tmp = open('{}{}.fa'.format(output_path,key),'w')
            f_tmp.write('>'+directory+'_'+key+'\n'+dictname[key]+'\n')
            f_tmp.close()
            
make_files('trop',tro,a)
make_files('chicken',chi,a)
make_files('mus',mus,a)
make_files('human',hum,a)
make_files('my',my,a)
```
